```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
hdb = pd.read_csv('../input/singapore-hdb-resale/HDBresale.csv')
hdb
```

```python
#total number of towns in Singapore
hdb['town'].nunique()
```

```python
# top 20 highest HDB resale_price in Singapore
hdb.sort_values(by='resale_price',ascending=False).iloc[:20]
```

```python
# percentage of transacted HDB property that cost more than 1 million dollars.
sum(hdb['resale_price'].apply(lambda x: x>=1000000))/len(hdb)*100
```

0.326% of transacted HDB properties costs more than 1 million.

```python
# percentage of transacted HDB property that cost between 900 thousand and 1 million
sum(hdb['resale_price'].apply(lambda x: 900000<=x<=1000000))/len(hdb)*100
```

0.985% of transacted HDB properties cost between 900 thousand and 1 million.

```python
# assuming that my budget is  $600,000
# percentage of transacted HDB property that is within our budget
sum(hdb['resale_price'].apply(lambda x: x<=600000))/len(hdb)*100
```

I am eligible to purchase about 85% of the properties with our budget.

```python
hdb.corr()
```

```python
#removing columns that i do not need
hdb1 = hdb.drop('block', axis=1).drop('street_name',axis=1).drop('flat_model',axis=1).drop('storey_range',axis=1)
hdb1
```

```python
# number of transaction of each type of flats from year 2017 onwards
hdb1.groupby('flat_type').count()['town']
```

The most popular flat type is 4 room.

```python
# number of transaction in each town from year 2017 onwards
transactions = hdb1.groupby('town').count()['resale_price']
transactions
```

```python
# the top 5 volumes of transactions from 2017 onwards
transactions.sort_values(ascending=False).iloc[:5]
```

OCR HDB flats has the highest number of transactions.

```python
# the lowest 5 volumes of transactions from 2017 onwards
transactions.sort_values(ascending=True).iloc[:5]
```

CCR and RCR flats has the lowest number of transactions.

```python
# number of transacted resale units with less than 50 years remaning lease
sum(hdb1['remaining_lease'].apply(lambda x: int(x[0:2])<=50))
```

```python
# percentage of transacted resale units with less than 50 years remaning lease
sum(hdb1['remaining_lease'].apply(lambda x: int(x[0:2])<=50))/len(hdb)*100
```

The number of transacted HDB flats for flats with less than 50 years lease is 2.02% which also indicates that people do not want to purchase flats that are too old.

So we should not purchase flats that are too old, it maybe tougher to sell in the future as the demand would be alot lower.
